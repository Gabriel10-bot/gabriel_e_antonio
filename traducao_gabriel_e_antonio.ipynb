```python
# Instale a biblioteca antes de rodar:
# pip install deep-translator

from deep_translator import GoogleTranslator

# Texto original
texto = "Hello, world! How are you today?"

# Tradução do inglês para o português
traducao = GoogleTranslator(source='en', target='pt').translate(texto)

print("Texto original:", texto)
print("Tradução:", traducao)
```

```python
# Instale a biblioteca antes de rodar:
# pip install deep-translator

from deep_translator import GoogleTranslator

# Texto original
texto = "Hello, world! How are you today?"

# Dicionário com os idiomas desejados (nome -> código ISO)
idiomas = {
    "Português": "pt",
    "Russo": "ru",
    "Coreano": "ko",
    "Inglês": "en",
    "Espanhol (México)": "es",  # O Google usa 'es' para espanhol geral
    "Italiano": "it"
}

print("Texto original:", texto)
print("-" * 50)

# Traduz para cada idioma da lista
for nome, codigo in idiomas.items():
    traducao = GoogleTranslator(source='auto', target=codigo).translate(texto)
    print(f"{nome}: {traducao}")
```

```python
from PIL import Image
import requests
from io import BytesIO
from transformers import pipeline

# Carregar modelo pré-treinado
classificador = pipeline("image-classification", model="microsoft/resnet-50")

# Imagem de exemplo
url_imagem = "https://upload.wikimedia.org/wikipedia/commons/9/9a/Pug_600.jpg"
imagem = Image.open(BytesIO(requests.get(url_imagem).content))
# imagem.show()  # opcional

# Classificar a imagem (corrigido!)
resultados = classificador(imagem)

print("Resultados da classificação:\n")
for r in resultados:
    print(f"{r['label']} - Confiança: {r['score']:.2%}")
```
